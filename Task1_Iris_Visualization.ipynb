```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Set the visual style for plots
sns.set_theme(style="whitegrid")
```

```python
# 1. Load the dataset
# Seaborn has the iris dataset built-in
iris = sns.load_dataset('iris')

# 2. Print shape and column names
print("shape:",iris.shape)
print("columns:",iris.columns.tolist())
print("first 5 rows:\n",iris.head())
print("dataset info:",iris.info())
print("dataset description:\n",iris.describe())
```

```python
# Scatter plot for Sepal features
sns.scatterplot(data=iris, x='sepal_length', y='sepal_width', hue='species')
plt.title('Sepal dimensions')
plt.show()

# Scatter plot for Petal features
sns.scatterplot(data=iris, x='petal_length', y='petal_width', hue='species')
plt.title('Petal dimensions')
plt.show()
```

```python
sns.pairplot(iris, hue='species')
plt.show()
```

```python
# Use histograms to show value distributions
iris.hist(edgecolor='black', linewidth=1.2, figsize=(12, 8))
plt.suptitle('Distribution of Features')
plt.show()
```

```python
# Use box plots to identify outliers
plt.figure(figsize=(12, 8))
plt.subplot(2,2,1)
sns.boxplot(x='species', y='sepal_length', data=iris)
plt.subplot(2,2,2)
sns.boxplot(x='species', y='sepal_width', data=iris)
plt.subplot(2,2,3)
sns.boxplot(x='species', y='petal_length', data=iris)
plt.subplot(2,2,4)
sns.boxplot(x='species', y='petal_width', data=iris)
plt.tight_layout()
plt.show()
```
